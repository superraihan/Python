Table Driven Agent

Table Driven Agent 

```python
table={
    ('A', 'Clean'): 'Right',
    ('A', 'Dirty'): 'Suck', 
    ('B', 'Clean'): 'Left',
    ('B', 'Dirty'): 'Suck'
}
```

```python
percept = []
def table_driven_agent(perc):
    print("Percept Read:", perc)
    percept.append(perc)
    action = table[perc]
    return action
```

Agent Simulation

```python
import random
location = random.choice(['A', 'B'])
condition = random.choice(['Clean', 'Dirty'])
action = table_driven_agent((location, condition))
print("Action Performed:" +action)
```

Simple Reflex Agent 

```python
def simple_reflex_agent(percept):
    print('percept Read:', percept)
    loc = percept[0]
    status = percept[1]
    if status =='Dirty':
        action =='Suck'
    elif loc =='A':
        action ='Right'
    elif loc =='B':
        action ='Left'  
    return action

```

Simulation

```python
import random

def simple_reflex_agent(percept):
    loc = percept[0]
    status = percept[1]
    
    if status == 'Dirty':
        action = 'Suck'
    elif loc == 'A':
        action = 'Right'
    else:
        action = 'Left'
    
    return action


# Simulation
location = random.choice(['A', 'B'])
condition = random.choice(['Clean', 'Dirty'])

while True:
    action = simple_reflex_agent((location, condition))
    print(f'Location: {location}, Condition: {condition}')
    print('Action Performed:', action)
    
    cmd = input('Continue? (y/n): ')
    if cmd.lower() != 'y':
        break
    
    if action == 'Right':
        location = 'B'
        condition = random.choice(['Clean', 'Dirty'])
    elif action == 'Left':
        location = 'A'
        condition = random.choice(['Clean', 'Dirty'])
    else:
        condition = 'Clean'
```

BFS Tree Search

```python
def bfs(startNode, goalNode,G):
    visited = []
    expand_list = []
    frontier = []
    snode ={'city': startNode, 'path Cost':0, 'path':[startNode]}
    frontier.append(snode)
    visited.append(startNode)
    while len(frontier)>0:
        print(frontier)
        unode = frontier.pop(0)
        u = unode['city']
        expand_list.append(u)
        print(u+" expanding.")

        for v in G[u].keys():
            print(v+" generated")
            cost = G[u][v]+ unode['path Cost']
            path = unode['path'] + [v]
            vnode = {'city': v, 'path Cost': cost, 'path': path}
            frontier.append(vnode)
            visited.append(v)
            if v == goalNode:
                print(expand_list)
                return vnode
            
```
