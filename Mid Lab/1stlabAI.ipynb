### string

```python
str1= "Hello"
str2= "Class"

#concate
message = str1 + " " + str2
print(message)
print(len(message))

#slicing
substr= message[2:7]
print("sub: ", substr)

#find substring
index = message.find("Class")
print("ind: ", index)

#replace
new_text = message.replace("Class", "Data Structure")
print("new: ", new_text)
print(len(new_text))
```

```python
#split
word= new_text.split()
print("word: ", word)
print(len(word))
text = "apple, banana, orange, grapes"
fruits = text.split(", ")
print("fruits: ", fruits)

#add new word to list
fruits.append("mango")
print(fruits)
print(fruits[0])
```

```python
fruits.insert(0, "berry")
fruits
```

```python
#only nunique values
print(set(fruits))
print(sorted(fruits))
fruits.sort()
print(fruits)
fruits.reverse()
print(fruits)       
print(fruits.count("berry"))
print(fruits[4].count("r"))
print(fruits.pop(2))
print(fruits)
```

### loop

```python
#starting from 0
for i in range(5):
    print(i, end=' ')
print()
#starting from other point
for i in range(2, 10):
    print(i, end=' ')
```

```python
#starting from other point and increment by 2
for i in range(2, 10, 2):
    print(i, end=' ')
print()
#decrement
for i in range(10, 0, -1):
    print(i, end=' ')
```

```python
#print only even number between range 0-20
for i in range(20):
    if i % 2 == 0:
        print(i, end=' ')
print()
#print only number that are multiple of 3
for i in range(20):
    if i % 3 == 0:
        print(i, end=' ')
```
