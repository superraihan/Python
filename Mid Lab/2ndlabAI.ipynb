#### TABLE dirven AGENT

```python
table = {
    ('A', 'Clean'): 'Right',
    ('A', 'Dirty'): 'Suck',
    ('B', 'Clean'): 'Left',
    ('B', 'Dirty'): 'Suck',
}
```

```python
percepts=[] #to store percept sequencer
def table_driven_agent(perc):
    print('Percept read:',perc)
    percepts.append(perc)
    action = table[perc]
    return action
```

```python
#agent simulation
import random
location = random.choice(['A','B'])
condition = random.choice(['Clean','Dirty'])
action = table_driven_agent((location, condition))
print('Action Performed:'+action)
```

#### Simple Reflex Agent

```python
def simple_reflex_agent(percept):
    print('Percept read:',percept)
    loc = percept[0]
    status = percept[1]
    if status == 'Dirty':
        action = 'Suck'
    elif loc == 'A':
        action = 'Right'
    elif loc == 'B':
        action = 'Left'
    return action
```

```python
#simulation
location = random.choice(['A','B'])
condition = random.choice(['Clean','Dirty'])
while True:
    action = simple_reflex_agent((location, condition))
    print('Action Performed:',action)
    cmd=input('continue?(y/n): ')
    if(cmd=='n' or cmd!='y'):
        break
    if action=='Right':
        location ='B'
        condition = random.choice(['Clean','Dirty'])
    elif action=='Left':
        location ='A'
        condition = random.choice(['Clean','Dirty'])
    else:
        condition='Clean'
```

#### MODEL BASE AGENT

```python
model = {'A': 'Unknown', 'B': 'Unknown'}
world_state='Bad'
action='null'
def update_state(action, location, status, model):
    global world_state, model,action
    model[location] = status
    if action == 'Suck':
        model[location] = 'Clean'
    if model['A'] == 'Clean' and model['B'] == 'Clean':
        world_state = 'Good'
    else:
        world_state = 'Bad'
    return world_state

def model_based_agent(percept):
    loc = percept[0]
    status = percept[1]
    global world_state, model
    if status == 'Dirty':
        action = 'Suck'
    elif loc == 'A':
        action = 'Right'
    elif loc == 'B':
        action = 'Left'
    elif world_state == 'Good':
        action = 'pause'
    world_state = update_state(loc, status)
    print('percepts',percept)
    print('Action:',action )
    print('model',model)
```
