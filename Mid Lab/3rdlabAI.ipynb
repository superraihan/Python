### BFS tree search

```python
def bfs (startNode, goalNode, G):
    visited = [] 
    expand_list = []
    frontier = []

    snode = {'City': startNode, 'Path Cost':0, 'Path':[startNode]}
    frontier.append(snode)
    visited.append(startNode)

    while len(frontier) > 0:
        print(frontier)
        unode=frontier.pop(0) # {'City': startNode, 'Path Cost':0, 'Path':[startNode]}
        u=unode['City']
        expand_list.append(u)
        print(u+" expanding.")

        for v in G[u].keys():
            print(v+" generated")
            cost= G[u][v] + unode['Path Cost']
            Path= unode['Path'] + [v]
            vnode={'City':v, 'Path Cost':cost, 'Path':Path}
            frontier.append(vnode)
            visited.append(v)

            if v == goalNode:
                print(expand_list)
                return vnode
                
```
