```python
import random
def initpopulation(b,n):
    p =[random.choices([0,1],k=b) for _ in range(n)]
    return p
```

```python
initpopulation(5,4)
```

```python
def binToDecimal(binList):
    decvalue = 0
    power = len(binList)
    for digit in binList:
        
        decvalue += digit *(2**(power-1))
        power -= 1
    return decvalue
def func(x):
    return x**2
def getFitness(p):
    f = []
    for sol in p:
        d = binToDecimal(sol)
        print('generated x:',d)
        v = func(d)
        if v<0: f.append(0)
        else: f.append(v)
    return f
```

```python
population = initpopulation(5,4)
getFitness(population)
```

```python
fitness = getFitness(population)
total_fitness = sum(fitness)
prob = [f/total_fitness for f in fitness]
n = len(population)
expected_counts = [n*pi for pi in prob]
print(fitness)
print(prob)
print(expected_counts)
```

```python
def mutation(p, pm):
    mutated = []
    for i in range(0, len(p), 2):
        c1 = p[i].copy()
        c2 = p[i+1].copy()
        
        for j in range(len(c1)):
            if random.random() < pm:
                c1[j] = 1 - c1[j]
            if random.random() < pm:
                c2[j] = 1 - c2[j]
        
        mutated.append(c1)
        mutated.append(c2)
    return mutated
```

```python
def get_individual_fitness(sol):
    d = binToDecimal(sol)
    v = func(d)
    return v if v >= 0 else 0

def selection(parent, mutated):
    selected = []
    for i in range(len(parent)):
        if get_individual_fitness(mutated[i]) >= get_individual_fitness(parent[i]):
            selected.append(mutated[i])
        else:
            selected.append(parent[i])
    return selected
```

```python
population = initpopulation(5,4)

offspring = cross(population)

if len(offspring) % 2 != 0:
    offspring = offspring[:-1]

mutated = mutation(offspring, 0.1)

# Select between original population and mutated offspring
limit = min(len(population), len(mutated))
new_population = selection(population[:limit], mutated[:limit])

print("Parent:", population)
print("Offspring:", offspring)
print("Mutated:", mutated)
print("Selected:", new_population)
```
